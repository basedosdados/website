## Migração Automática

```python
class Migrator:
    def __init__(self, ckan_local: RemoteCKAN, ckan_remote: RemoteCKAN, folder: str):
        self.ckan_local = ckan_local
        self.ckan_remote = ckan_remote
        self.datasets = self.load_filepath(folder)

        if not Path("./ckan-original-stage").exists():
            Path("./ckan-original-stage").mkdir()

        if not Path("./ckan-updated").exists():
            Path("./ckan-updated").mkdir()

    ###########################################################

    def migrate_all(self, force_create=False):
        for index in range(len(self.datasets)):
            self.migrate_one(index, force_create)
            clear_output(wait=True)

    def migrate_one(self, index, force_create=False):
        print(index)
        print(self.datasets[index].parent.name)
        dataset = self.load_dataset_updated(index)
        if force_create:
            if dataset.get("id"):
                del dataset["id"]
            for i, resource in enumerate(dataset.get("resources") or []):
                if resource.get("id"):
                    del dataset["resources"][i]["id"]
            return self.ckan_local.action.package_create(**dataset)
        if dataset.get("id"):
            # se tem id, então é dataset do ckan
            return self.ckan_local.action.package_update(**dataset)
        else:
            # senão é um dataset novo do github
            return self.ckan_local.action.package_create(**dataset)

    ###########################################################

    def validate_all(self):
        for index in range(len(self.datasets)):
            self.validate_one(index)
            clear_output(wait=True)

    def validate_one(self, index):
        print(index)
        print(self.datasets[index].parent.name)
        dataset = self.load_dataset_updated(index)
        return self.ckan_local.action.package_validate(**dataset)

    ###########################################################

    def save_all_migrated(self):
        for index in range(len(self.datasets)):
            self.save_one_migrated(index)
            clear_output(wait=True)

    def save_one_migrated(self, index):
        updates = self.load_dataset_original(index)
        dataset_id = updates["dataset_id"].replace("_", "-")
        dataset_filepath = Path(f"./ckan-migrated/{dataset_id}.json")

        try:
            current = self.ckan_local.action.package_show(id=dataset_id)
            with open(dataset_filepath, "w") as file:
                json.dump(current, file)
        except:
            print(f"ERROR: Dataset {dataset_id} not found")
            current = {}

    ###########################################################

    def save_all_updated(self):
        for index in range(len(self.datasets)):
            self.save_one_updated(index)
            clear_output(wait=True)

    def save_one_updated(self, index):
        dataset = self.load_dataset_updated(index)
        if not dataset.get("id"):
            dataset["owner_org"] = "3626e93d-165f-42b8-bde1-2e0972079694"
        with open(f"./ckan-updated/{dataset['name']}.json", "w") as file:
            json.dump(dataset, file)

    ###########################################################

    def save_all_ckan(self):
        for index in range(len(self.datasets)):
            self.save_one_ckan(index)
            clear_output(wait=True)

    def save_one_ckan(self, index):
        updates = self.load_dataset_original(index)
        dataset_id = updates["dataset_id"].replace("_", "-")
        dataset_filepath = Path(f"./ckan-original-stage/{dataset_id}.json")

        try:
            current = self.ckan_remote.action.package_show(id=dataset_id)
            with open(dataset_filepath, "w") as file:
                json.dump(current, file)
        except:
            print(f"ERROR: Dataset {dataset_id} not found")
            current = {}

    ###########################################################

    def load_dataset_updated(self, index):
        dataset_id = self.datasets[index].parent.name
        dataset_id = dataset_id.replace("_", "-")
        dataset_filepath = Path(f"./ckan-updated/{dataset_id}.json")

        if dataset_filepath.exists():
            with open(dataset_filepath, "r") as file:
                return json.load(file)

        dataset = self.load_dataset_original(index)
        dataset = self.parse_dataset(dataset)

        return dataset

    def load_dataset_original(self, index):
        dataset = self.datasets[index]
        dataset = dataset.open("r").read()
        dataset = yaml.safe_load(dataset)

        resources = self.datasets[index].parent
        resources = resources.glob("**/table_config.yaml")

        if "resources" not in dataset:
            dataset["resources"] = []

        for resource in resources:
            resource = resource.open("r").read()
            resource = yaml.safe_load(resource)
            dataset["resources"].append(resource)

        dataset = self.omit_hints(dataset)

        print(f"{'-' * 60}")
        print(f"{index:03} {'-' * 56}")

        return dataset

    ###########################################################

    def parse_dataset(self, updates):
        ###########################################################
        # Get local dataset #######################################
        ###########################################################
        print(f"LOG: Dataset {updates['dataset_id']}")
        print(f"LOG: Organization {updates['organization']}")

        dataset_id = updates["dataset_id"].replace("_", "-")
        dataset_filepath = Path(f"./ckan-original-stage/{dataset_id}.json")

        if dataset_filepath.exists():
            current = json.load(dataset_filepath.open("r"))
        else:
            print(f"ERROR: Dataset {dataset_id} not found")
            current = {
                "name": dataset_id,
                "notes": "",
                "owner_org": owner_org_mapping[updates["organization"]],
                "resources": [],
                "groups": [],
                "tags": [],
            }
        ###########################################################
        # Format dataset metadata #################################
        ###########################################################
        current["notes"] += updates.get("description")
        current["author"] = (
            updates.get("author").get("name") if "author" in updates else None
        )
        current["author_email"] = (
            updates.get("author").get("email") if "author" in updates else None
        )
        current["state"] = "active"
        if updates.get("license").get("name"):
            current["license_id"] = ""
            current["license_title"] = updates["license"]["name"]
        current["url"] = updates.get("url") or current.get("url")
        current["private"] = False
        if current.get("organization"):
            if current.get("organization").get("name") != updates.get("organization"):
                print(f"ERROR: Organization mismatch")
                print(f"> Current organization: {current['organization']['name']}")
                print(f"> Updated organization: {updates['organization']}")
        else:
            current["organization"] = {"name": updates["organization"]}
        current["ckan_url"] = updates["url_ckan"]
        current["github_url"] = updates["url_github"]
        ###########################################################
        # Extras ##################################################
        ###########################################################
        current_extras = [extras.get("key") for extras in current.get("extras") or []]
        if "dataset_args" in current_extras:
            key = current_extras.index("dataset_args")
            value = (
                ast.literal_eval(current["extras"][key]["value"])
                if isinstance(current["extras"][key]["value"], str)
                else current["extras"][key]["value"]
            )
            value["dataset_id"] = updates.get("dataset_id")
            current["extras"][key]["value"] = value
        else:
            current["extras"] = [
                {
                    "key": "dataset_args",
                    "value": {"dataset_id": updates.get("dataset_id")},
                }
            ]
        current = self.delete_dataset_extra_fields(current)
        ###########################################################
        # Format bdm table metadata ###############################
        ###########################################################
        current_bdm_tables = map(lambda x: x.get("name"), current["resources"])
        current_bdm_tables = list(current_bdm_tables)
        for resource in updates["resources"]:
            resource_key = resource["table_id"]
            if resource_key in current_bdm_tables:
                index = current_bdm_tables.index(resource_key)
                current["resources"][index] = self.parse_bdm_table(
                    current["resources"][index], resource
                )
            else:
                new_resource = self.parse_bdm_table({}, resource)
                current["resources"].append(new_resource)
        for i, resource in enumerate(current["resources"]):
            if resource["resource_type"] == "bdm_table":
                current["resources"][i] = self.delete_bdm_table_extra_fields(
                    current["resources"][i]
                )
        ###########################################################
        # Format external links metadata ##########################
        ###########################################################
        current_external_links = map(lambda x: x.get("url"), current["resources"])
        current_external_links = list(current_external_links)
        for url in updates.get("website") or []:
            if url in current_external_links:
                index = current_external_links.index(url)
                current["resources"][index] = self.parse_external_link(
                    current["resources"][index], updates, url, False
                )
            else:
                new_resource = self.parse_external_link({}, updates, url, True)
                current["resources"].append(new_resource)
        for index, resource in enumerate(current["resources"]):
            if current["resources"][index].get("temporal_coverage"):
                current["resources"][index]["temporal_coverage"] = [
                    value
                    for value in current["resources"][index]["temporal_coverage"]
                    if isinstance(value, int)
                ]
        for i, resource in enumerate(current["resources"]):
            if resource["resource_type"] == "external_link":
                current["resources"][i] = self.delete_external_link_extra_fields(
                    current["resources"][i]
                )
        ###########################################################
        # Format groups metadata ##################################
        ###########################################################
        groups = map(lambda x: x.get("name"), current.get("groups", []))
        groups = list(groups)
        for group in updates.get("groups", []) or []:
            if group not in groups:
                current["groups"].append({"name": group})
        ###########################################################
        # Format tags metadata ####################################
        ###########################################################
        tags = map(lambda x: x.get("name"), current.get("tags", []))
        tags = list(tags)
        for tag in updates.get("tags", []) or []:
            if tag not in tags:
                current["tags"].append({"name": tag})
        return current

    def parse_bdm_table(self, current, updates):
        current["name"] = updates.get("table_id")
        current["description"] = updates.get("description")
        current["url"] = ""
        current["resource_type"] = "bdm_table"
        current["dataset_id"] = updates.get("dataset_id")
        current["table_id"] = updates.get("table_id")
        current["spatial_coverage"] = ",".join(
            [
                spatial_coverage_mapping.get(key, key)
                for key in (updates.get("coverage_geo", []) or [])
            ]
        )
        current["temporal_coverage"] = (
            [t for t in updates.get("coverage_time", []) if isinstance(t, int)]
            if updates.get("coverage_time")
            else None
        )
        current["update_frequency"] = update_frequency_mapping.get(
            updates.get("data_update_frequency")
        )
        current["entity"] = [
            entity_mapping[key] for key in (updates.get("observation_level") or [])
        ]
        current["entity"] = filter(lambda x: x, current["entity"])
        current["entity"] = list(current["entity"])
        current["identifying_columns"] = updates.get("primary_keys", None)
        current["version"] = str(updates.get("version") or "v1")
        current["published_by"] = {
            "name": (updates.get("published_by") or {}).get("name"),
            "email": (updates.get("published_by") or {}).get("email"),
            "website": (updates.get("published_by") or {}).get("website"),
            "ckan_user": (updates.get("published_by") or {}).get("ckan_user"),
            "github_user": (updates.get("published_by") or {}).get("github_user"),
        }
        current["data_cleaned_by"] = (
            {
                "name": (updates.get("treated_by") or {}).get("name"),
                "email": (updates.get("treated_by") or {}).get("email"),
                "website": (updates.get("treated_by") or {}).get("website"),
                "code_url": (updates.get("treated_by") or {}).get("code_url"),
                "ckan_user": (updates.get("treated_by") or {}).get("ckan_user"),
                "github_user": (updates.get("treated_by") or {}).get("github_user"),
            }
            if "treated_by" in updates
            else None
        )
        current["data_cleaning_description"] = updates.get("treatment_description")
        if isinstance(current["data_cleaning_description"], list):
            current["data_cleaning_description"] = " ".join(
                current["data_cleaning_description"]
            )
        current["source_bucket_name"] = updates.get("source_bucket_name")
        current["project_id_prod"] = updates.get("project_id_prod")
        current["project_id_staging"] = updates.get("project_id_staging")
        current["partitions"] = updates.get("partitions") or ""
        if isinstance(current["partitions"], list):
            current["partitions"] = ", ".join(current["partitions"])
        current["columns"] = updates.get("columns")

        return current

    def parse_external_link(self, current, updates, url, change_name):
        current["resource_type"] = "external_link"
        current["name"] = "Visualizar" if change_name else current.get("name")
        current["dataset_id"] = updates["dataset_id"]
        current["url"] = url
        current["language"] = [
            language_mapping.get(l) for l in updates.get("languages") or []
        ]
        current["has_api"] = "yes" if updates.get("API") == "sim" else "no"
        current["is_free"] = "yes" if updates.get("free") == "sim" else "no"
        current["requires_registration"] = (
            "yes" if updates.get("registration") == "sim" else "no"
        )
        current["availability"] = updates.get("availability")
        current["country_ip_address_required"] = (
            ["bra"] if updates.get("brazilian_IP") == "sim" else None
        )
        if updates.get("license").get("name") == "MIT":
            current["license"] = "mit"
        elif "license" in current:
            del current["license"]
        return current

    def delete_dataset_extra_fields(self, dataset):
        if "num_resources" in dataset:
            del dataset["num_resources"]
        if "num_tags" in dataset:
            del dataset["num_tags"]
        if "license_url" in dataset:
            del dataset["license_url"]
        dataset.pop("download_type", None)
        dataset.pop("dataset_id", None)
        return dataset

    def delete_bdm_table_extra_fields(self, resource):
        if "coverage_geo" in resource:
            del resource["coverage_geo"]
        if "coverage_time" in resource:
            del resource["coverage_time"]
        if "data_update_frequency" in resource:
            del resource["data_update_frequency"]
        if "entity" in resource and resource["entity"] and len(resource["entity"]) == 0:
            del resource["entity"]
        if "observation_level" in resource:
            del resource["observation_level"]
        if "primary_keys" in resource:
            del resource["primary_keys"]
        if "publisher" in resource:
            del resource["publisher"]
        if "publisher_email" in resource:
            del resource["publisher_email"]
        if "publisher_github" in resource:
            del resource["publisher_github"]
        if "publisher_website" in resource:
            del resource["publisher_website"]
        if "treatment_description" in resource:
            del resource["treatment_description"]
        if "published_by.code_url" in resource:
            del resource["published_by.code_url"]
        if "published_by.email" in resource:
            del resource["published_by.email"]
        if "published_by.name" in resource:
            del resource["published_by.name"]
        if "published_by.website" in resource:
            del resource["published_by.website"]
        if "publisher" in resource:
            del resource["publisher"]
        if "publisher_email" in resource:
            del resource["publisher_email"]
        if "publisher_github" in resource:
            del resource["publisher_github"]
        if "publisher_website" in resource:
            del resource["publisher_website"]
        if "treated_by.code_url" in resource:
            del resource["treated_by.code_url"]
        if "treated_by.email" in resource:
            del resource["treated_by.email"]
        if "treated_by.name" in resource:
            del resource["treated_by.name"]
        if "treated_by.website" in resource:
            del resource["treated_by.website"]
        if "treatment_description" in resource:
            del resource["treatment_description"]
        if "url_ckan" in resource:
            del resource["url_ckan"]
        if "url_github" in resource:
            del resource["url_github"]

        return resource

    def delete_external_link_extra_fields(self, resource):
        if "brazilian_ip" in resource:
            del resource["brazilian_ip"]
        if "free" in resource:
            del resource["free"]
        if "license_type" in resource:
            del resource["license_type"]
        if "signup_needed" in resource:
            del resource["signup_needed"]
        if "version" in resource:
            del resource["version"]
        if "bdm_file_size" in resource:
            del resource["bdm_file_size"]
        return resource

    ###########################################################

    def load_filepath(self, folder):
        datasets = Path(folder).glob("**/**/dataset_config.yaml")
        datasets = list(datasets)
        datasets = sorted(datasets)
        return datasets

    def omit_hints(self, data):
        """Set values encapsulated with <> to None in a dict"""
        if isinstance(data, str):
            hidden = f"{data[0]}{data[-1]}" != "<>" if len(data) >= 2 else None
            return data if hidden else None
        elif isinstance(data, list):
            lst = list(filter(self.omit_hints, data))
            return lst if lst else None
        elif isinstance(data, dict):
            for key in data.keys():
                data[key] = self.omit_hints(data[key])
            return data
        return data


###########################################################
###########################################################
###########################################################
###########################################################

ckan_local = RemoteCKAN(
    "https://staging.basedosdados.org/",
    "eyJ0eXAiOiJKV1QiLCJhbGciOiJIUzI1NiJ9.eyJqdGkiOiI5ZUZ6eUJHdWZIYUI3NjEzbW9kTEhRY3hQbG9Ja0dSS3Rzb1dWUnRHRTR1WlBELW1OYng1Tm1qVFQtRGdTdnFCR2wzR0V0emllTFZlRndVMiIsImlhdCI6MTYyOTE1NjM1OX0.Vh4EGl997WA3VljIsty_jn_onNP-gPxMIsqy8ECYsg0",
)

ckan_remote = RemoteCKAN("https://staging.basedosdados.org/")

migrator = Migrator(ckan_local, ckan_remote, "mais/bases")
```
